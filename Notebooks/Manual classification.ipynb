```python
import os, re, random, string
from collections import defaultdict
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
import numpy as np
from matplotlib.dates import date2num
import pytz
import pickle
```

```python
# Get the given feature from the file header
# Features: Message-ID, Date, From, To, Subject, Mime-Version, Content-Type, Content-Transfer-Encoding, 
# X-From, X-To, X-cc, X-bcc, X-Folder, X-Origin, X-FileName
def getFeature(feature, email):
    if feature == 'To':
        contains = re.findall(feature + ': (.*?)\nSubject', email, re.DOTALL)
    else:
        contains = re.findall(feature + ': (.*?)\n', email) # get text between colon and newline
    found = ''
    if contains:
        return contains[0]
    else: return ''
```

```python
# Get the contents of the email
def getBody(email):
    contains = re.search('X-FileName: (.*?)\Z', email, re.DOTALL) # text between last line of file header & end of file
    found = ''
    if contains:
        found = contains.group(1)
        found = found.split('\n',1)[-1] # remove final line of header from text
    return found[1:]
```

```python
# read the file
def getFileContents(email):
    f = open(email, 'r', encoding='latin1')
    contents = f.read()
    f.close()
    return contents
```

```python
# Making a list of file paths for all emails

emails = list()
for root, dirs, files in os.walk("./maildir", topdown = True):
    for name in files:
        if not name.startswith('.'): # files which start with . are not emails
            emails.append(os.path.join(root, name))
```

```python
testEmails = random.sample(emails, 1)
for email in testEmails:
    file_contents = getFileContents(email)
    body = getBody(file_contents)
    emailLines = body.splitlines()
    print(body)
    lineno = 1
    for line in emailLines:
        emailID = email + 'lineno' + str(lineno)
        print(emailID)
        lineno += 1
```

```python
# g = greeting
# b = body
# se = section separator
# so = signoff
# sa = signature
# a = attachment
# th = thread header
# tg = greeting within thread
# tb = body within thread
# tso = signoff within thread
# tsa = signature within thread
```

```python
for email in sampleEmails:
    file_contents = getFileContents(email)
    body = getBody(file_contents)
    emailLines = body.splitlines()
    lineno = 1
    for line in emailLines:
        print(line)
        lineClass = lineClasses[email + 'lineno' + str(lineno)]
        print('This line is a ' + classToText(lineClass))
        lineno += 1
```

```python
def classToText(lineClass):
    if lineClass == 'g':
        return 'greeting'
    if lineClass == 'b':
        return 'body'
    if lineClass == 'se':
        return 'separator'
    if lineClass == 'so':
        return 'signoff'
    if lineClass == 'sa':
        return 'signature'
    if lineClass == 'a':
        return 'attachment'
    if lineClass == 'th':
        return 'thread header'
    if lineClass == 'tg':
        return 'thread greeting'
    if lineClass == 'tb':
        return 'thread body'
    if lineClass == 'tso':
        return 'thread signoff'
    return 'thread signature'
```
